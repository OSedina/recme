```python
from collections import Counter
from collections.abc import Iterable


class SeriesFreq:
    def __init__(self):
        self.freq = Counter()
        
    def fit(self, iterable):
        for i in iterable:
            if isinstance(i, Iterable):
                self.freq.update(i)
        return self
    
    def to_series(self):
        return pd.Series(self.freq)
    
    def sort(self, iterable):
        if isinstance(iterable, Iterable):
            return sorted(iterable, key=lambda x: self.freq[x], reverse=True)
```

```python
sf_viewed, sf_bought = SeriesFreq().fit(full_train.iloc[:, 0]), SeriesFreq().fit(full_train.iloc[:, 1])
```

```python
def precision(session, k=1):
    count = 0
    if session[1]:
        for i in session[0][:k]:
            if i in session[1]:
                count += 1
    return count / k


def recall(session, k=1):
    count = 0
    if session[1]:
        for i in session[0][:k]:
            if i in session[1]:
                count += 1
        return count / len(session[1])
    return 0
```

```python
round(full_train.apply(lambda x: precision(x), axis=1).mean(), 2)
```

```python
round(full_train.apply(lambda x: precision(x, 5), axis=1).mean(), 2)
```

```python
round(full_train.apply(lambda x: recall(x), axis=1).mean(), 2)
```
